```python
from bs4 import BeautifulSoup
from time import sleep
from random import randint
from requests import get
import pandas as pd
```

```python
headers = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'})
```

```python
url= 'https://www.niche.com/k12/search/best-school-districts/m/st-louis-metro-area/'
        
r=get(url,headers=headers)
soup = BeautifulSoup(r.text, 'html.parser')
card = soup.find_all('div',class_='search-result')
dist_urls=[]
```

```python
for x in card:
    try: 
        rank_title=(x.find('div', class_="search-result-badge")).text
        print(rank_title)
    except AttributeError:
        print("Nontype text")
    School_district= x.find('h2').text
    print(School_district)
    grade=(x.find('figure', class_="search-result-grade")).text
    print(grade[:2])
    dist_url=(x.find('a')['href'])
    print(dist_url)
    re=get(dist_url,headers=headers)
    soup = BeautifulSoup(re.text, 'html.parser')  
    #schools=soup.find("li",class="tabbed-content-tab")
    zipcode= soup.find("div", class_="profile__address").text
    zipcode= zipcode[-6:]
    print(zipcode)
    

        
```
