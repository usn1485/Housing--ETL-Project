```python
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
from bs4 import BeautifulSoup
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.usnews.com/education/best-high-schools/missouri/rankings'
browser.visit(url)

```

```python
done = False

while done == False:

    try:
        browser.click_link_by_partial_text('Load More')
          
    except:
        print("Page Load Complete")
        done = True
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

school_district_fields = soup.find_all('div', class_='search-result-content')

district_name_list = []


for school_districts in school_district_fields:
    
    # Get names
    h3 = school_districts.find('h3')
    district_name = h3.text.strip()
    district_name_list.append(district_name)
    print(district_name)
    
    
try:
    browser.click_link_by_partial_text('Load More')
          
except:
    print("Scraping Complete")
    
    
    
    



#for district in districts:
#print(school_district)
#print(school_district_fields)
```
